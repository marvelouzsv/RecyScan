**Explore Data "Link harga sampah plastik"**

Download dataset pada link berikut: https://drive.google.com/file/d/1y9Z96sMrbgu6740SiTLbkbhAEifACUJj/view?usp=drive_link

```python
# Define the paths
dir_path = '/content/drive/My Drive/Bangkit/'  # Replace with the actual path to the directory containing your CSV file in Google Drive
file_path = os.path.join(dir_path, 'Harga Sampah Plastik.csv')  # Replace with your actual file name
```

```python
# Step 2: Read the CSV file into a DataFrame
price_df = pd.read_csv(file_path)
```

```python
# Step 3: Display the DataFrame
print("The dataframe: ")
display(price_df)
print()
```

```python
# Step 4: Display the data types of each column in the DataFrame
print("The datatypes: ")
print(price_df.dtypes)
print()
```

```python
# Step 5: Mencetak "Harga sebelum" dan "Harga sesudah" untuk botol
print("Harga sebelum daur ulang:", price_df['Harga Sebelum Daur Ulang'][0], "| Harga setelah daur ulang:", price_df['Harga Setelah Daur Ulang'][0])
```

**Merge Data "Link Rekomendasi Daur Ulang Sampah" and "Harga Sampah Plastik"**

```python
link_price_df = pd.merge(
    left=link_df,
    right=price_df,
    how="left",
    left_on="Jenis Sampah Plastik",
    right_on="Jenis Sampah Plastik"
)
link_price_df.head()
```

```python
link_price_df = link_price_df.drop(columns=['Sebelum', 'Setelah'])
```

```python
link_price_df.head()
```

```python
# Mencetak "Harga sebelum" dan "Harga sesudah" untuk botol
print("Rekomendasi Cara Daur ulang untuk", link_price_df['Jenis Sampah Plastik'][0], "adalah:", link_price_df['Link Rekomendasi'][0])
print("Harga sebelum daur ulang:", link_price_df['Harga Sebelum Daur Ulang'][0], "| Harga setelah daur ulang:", link_price_df['Harga Setelah Daur Ulang'][0])
```
