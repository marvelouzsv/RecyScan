```python
from google.colab import drive
import zipfile
import os
from matplotlib import image, pyplot as plt
```

```python
# Step 1: Mount Google Drive
drive.mount('/content/drive')
```

```python
# Define the paths
zip_file_path = '/content/drive/My Drive/Bangkit/plastic_waste_new.zip'  # Replace with the actual path to your ZIP file in Google Drive
destination_dir = '/content/data/'  # Path where you want to extract the files

# Create the destination directory if it doesn't exist
os.makedirs(destination_dir, exist_ok=True)
```

```python
# Step 2: Extract the contents of the ZIP file to the destination directory
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(destination_dir)

base_path = '/content/data/plastic/'
```

```python
# Step 3: List labels and analyze images
if os.path.exists(base_path):
    labels = os.listdir(base_path)
    print("Labels:", labels)
else:
    print(f"The directory {base_path} does not exist.")
```

```python
# Initialize the total image counter
total_images = 0

# Iterate through each label and count the number of images
for label in labels:
    directory = os.path.join(base_path, label)
    num_images = len(os.listdir(directory))
    total_images += num_images
    print(num_images, "\t" + label)

# Print the total number of images
print("Total images:", total_images)
```

```python
for label in labels:
    directory = os.path.join(base_path, label)
    path = os.path.join(directory, os.listdir(directory)[0])
    i = image.imread(path)
    print(i.shape, "\t" + label)
    # Optional: Display the image
    plt.imshow(i)
    plt.title(label)
    plt.show()
```
